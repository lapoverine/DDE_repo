Колонки "Academic Pressure", "Work Pressure", "Study Satisfaction", "Study Satisfaction" имеют большое количество пропущенных значений, однако если присмотреться, то можно понять, что эти признаки связаны между собой (то есть тот кто работает, у того нет Academic Pressure, и наоборот, кто только учится - у него отсутствует Job Satisfaction). Данную взаимосвязь можно использовать для преобразования этих фичей, чтобы в дальнейшем было легче работать (например, если подавать этот датасет для обучения моделей)

```python
# Пропуски по колонкам
print(data.isnull().sum())
```

```python
# Сумма пропусков по колонкам
missing_data = data.isnull().sum()
# Процент пропусков
missing_percent = (missing_data[missing_data > 0] / data.shape[0]) * 100
# Визуализация (heatmap)

sns.heatmap(data.isnull(), cbar=False, yticklabels=False)
```

Зная, что эти колонки взаимосвязаны между собой, можно поработать над созданием комбинированных признаков и флагов. Для CGPA, которое определяется только для студентов, у рабочих людей можно заполнить специфическим значением (например, -1)

```python
data['Overall_Pressure'] = data['Academic Pressure'].fillna(0) + data['Work Pressure'].fillna(0)
data['Overall_Satisfaction'] = data['Study Satisfaction'].fillna(0) + data['Job Satisfaction'].fillna(0)

data['Has_Academic_Data'] = (~data['Academic Pressure'].isna()).astype(int)
data['Has_Work_Data'] = (~data['Work Pressure'].isna()).astype(int)
data['Has_CGPA'] = (~data['CGPA'].isna()).astype(int)
data.loc[data['Working Professional or Student'] == 'Working Professional', 'CGPA'] = -1

data = data.drop(['Academic Pressure', 'Work Pressure'], axis=1)
data
```
