# Анализ пропусков

Колонки "Academic Pressure", "Work Pressure", "Study Satisfaction", "Study Satisfaction" имеют большое количество пропущенных значений, однако если присмотреться, то можно понять, что эти признаки связаны между собой (то есть тот кто работает, у того нет Academic Pressure, и наоборот, кто только учится - у него отсутствует Job Satisfaction). Данную взаимосвязь можно использовать для преобразования этих фичей, чтобы в дальнейшем было легче работать (например, если подавать этот датасет для обучения моделей)

```python
# Пропуски по колонкам
print(data.isnull().sum())
```

```python
missing_data = data.isnull().sum()
missing_percent = (missing_data[missing_data > 0] / data.shape[0]) * 100
sns.heatmap(data.isnull(), cbar=False, yticklabels=False)
```

Зная, что эти колонки взаимосвязаны между собой, можно поработать над созданием комбинированных признаков и флагов. Для CGPA, которое определяется только для студентов, у рабочих людей можно заполнить специфическим значением (например, -1)

```python
data['Overall_Pressure'] = data['Academic Pressure'].fillna(0) + data['Work Pressure'].fillna(0)
data['Overall_Satisfaction'] = data['Study Satisfaction'].fillna(0) + data['Job Satisfaction'].fillna(0)

data['Has_Academic_Data'] = (~data['Academic Pressure'].isna()).astype(int)
data['Has_Work_Data'] = (~data['Work Pressure'].isna()).astype(int)
data['Has_CGPA'] = (~data['CGPA'].isna()).astype(int)
data.loc[data['Working Professional or Student'] == 'Working Professional', 'CGPA'] = -1

#data = data.drop(['Academic Pressure', 'Work Pressure', 'Study Satisfaction', 'Job Satisfaction'], axis=1)

data
```

# Оценка целостности и полноты

Рассчитаем метрики качества данных (Completeness и Uniqueness)

```python
print("1. COMPLETENESS - Полнота данных (≥ 0.98)")
print("="*50)

completeness_results = {}
completeness_threshold = 0.98

for column in data.columns:
    total_rows = len(data)
    non_null_count = data[column].count()
    completeness_ratio = non_null_count / total_rows
    
    completeness_results[column] = {
        'ratio': completeness_ratio,
        'passed': completeness_ratio >= completeness_threshold
    }
    
    status = "PASS" if completeness_ratio >= completeness_threshold else "FAIL"
    print(f"{column:15} | {completeness_ratio:.4f} | {status}")

completeness_passed = sum(1 for r in completeness_results.values() if r['passed'])
completeness_total = len(completeness_results)
print(f"\nИтог: {completeness_passed}/{completeness_total} колонок прошли проверку")
```

```python
print("\n2. UNIQUENESS - Уникальность данных (= 1.0)")
print("="*50)

# Колонки, значения в которых должны быть уникальными. В нашем случае - только колонка id
identifier_columns = ['id']

uniqueness_results = {}
uniqueness_threshold = 1.0

for column in identifier_columns:
    if column in data.columns:
        total_rows = len(data)
        unique_count = data[column].nunique()
        uniqueness_ratio = unique_count / total_rows
        duplicate_count = total_rows - unique_count
        
        uniqueness_results[column] = {
            'ratio': uniqueness_ratio,
            'passed': uniqueness_ratio == uniqueness_threshold,
            'duplicates': duplicate_count
        }
        
        status = "PASS" if uniqueness_ratio == uniqueness_threshold else "FAIL"
        print(f"{column:15} | {uniqueness_ratio:.4f} | Дубликатов: {duplicate_count} | {status}")

uniqueness_passed = sum(1 for r in uniqueness_results.values() if r['passed'])
uniqueness_total = len(uniqueness_results)
print(f"\nИтог: {uniqueness_passed}/{uniqueness_total} колонок прошли проверку")
```
